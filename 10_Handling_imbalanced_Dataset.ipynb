Imbalanced Dataset:
  
  A dataset with an unequal class distribution


```python
#  importing the dependencies
import numpy as np
import pandas as pd
```

```python
# loading the dataset to pandas DataFrame
credit_dataset = pd.read_csv('/content/credit_data.csv')
```

```python
# printing first 5 rows of dataframe
credit_dataset.head()
```

```python
credit_dataset.tail()
```

```python
# distribution of the two classes
credit_dataset['Class'].value_counts()
```

This is Highly imbalanced Dataset

0--------------->Legit Transactions

1--------------->Fraud Transactions

```python
# Separating the legit and fraudlent Transactions
legit = credit_dataset[credit_dataset.Class == 0.0]
fraud = credit_dataset[credit_dataset.Class == 1.0 ]
```

```python
print(legit.shape)
print(fraud.shape)
```

Under-Sampling

Build a sample dataset containing similar distribution of legit and fraudulent Transactions

Number of Fraudlent Transaction -->148

```python
legit_sample = legit.sample(n=148)
```

```python
print(legit_sample.shape)
```

Concatenate the Two DataFrames

```python
new_dataset = pd.concat([legit_sample,fraud],axis=0)
```

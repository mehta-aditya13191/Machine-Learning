Label Encodign:

   .Converting the labels into numeric form

```python
# importing the Dependencies
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

Label Encoding of Breast Cancer Dataset

```python
#  loading the data from csv file to panda dataframe
cancer_data=pd.read_csv('/content/breast_cancer_data.csv')
```

```python
cancer_data.head()
```

```python
# finding the count of differents labels
cancer_data['diagnosis'].value_counts()
```

```python
# load the LabelEncoder function
label_encoder=LabelEncoder()
```

```python
labels = label_encoder.fit_transform(cancer_data.diagnosis)
```

```python
# appending the labels to the DataFrame
cancer_data['target']=labels
```

```python
cancer_data.head()
```

0 -------->Benign

1--------->Malignant

```python
cancer_data['target'].value_counts()
```

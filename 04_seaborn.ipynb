Seaborn:
  
  . Data Visualization Library

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

Note : Seaborn has some built-in datasets

```python
# total bill vs tip dataset
tips=sns.load_dataset('tips')
```

```python
tips.head()
```

```python
# visualize the data
sns.set_theme()
sns.relplot(data=tips,x='total_bill',y='tip',col='time',hue='smoker',style='smoker',size='size')
```

```python
# load the iris dataset
iris=sns.load_dataset('iris')
```

```python
iris.head()
```

Scatter plot

```python
sns.scatterplot(x='sepal_length',y='petal_length',hue='species',data=iris)
```

```python
sns.scatterplot(x='sepal_length',y='petal_width',hue='species',data=iris)
```

```python
# house price dataset
from sklearn.datasets import load_boston
house_boston=load_boston()
```

```python
house_boston
```

```python
house=pd.DataFrame(data=house_boston.data,columns=house_boston.feature_names)
```

```python
house.head()
```

```python
house['price']=house_boston.target
```

```python
house.head()
```

Distribution Plot

```python
sns.distplot(house['price'])
```

Correlation Matrix

HEAT MAP

```python
correlation=house.corr()
```

```python
# constructin a Heat Map
plt.figure(figsize=(10,10))
sns.heatmap(correlation,cbar=True,square=True,fmt='.1f',annot=True,annot_kws={'size':8},cmap='Blues')
```
